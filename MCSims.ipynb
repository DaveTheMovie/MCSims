```python
# Simulation of value at T
def MC(S0,T,r,s,dt = 1/252):
    St = S0
    n = int(round(T/dt,0))
    r_dt = r * dt
    s_sqrt_dt = s * np.sqrt(dt)
    for i in range(n):
        dz = np.random.normal(0,1)
        St = St * (1+(r_dt + s_sqrt_dt * dz))
    return St
```

```python
# Simulation of vector
def MC(S0,T,r,s,dt = 1/252):
    St = S0
    n = int(round(T/dt,0))
    r_dt = r * dt
    s_sqrt_dt = s * np.sqrt(dt)
    sims = np.zeros(n+1)
    sims[0] = S0
    for i in range(n):
        dz = np.random.normal(0,1)
        sims[i+1] = sims[i] * (1+(r_dt + s_sqrt_dt * dz))
    return sims    
```

```python
# Efficient Simulation of vector
def MCT(S0,T,r,s,dt = 1/252):
    St = S0
    n = int(round(T/dt,0))
    r_dt = r * dt
    s_sqrt_dt = s * np.sqrt(dt)
    sims = np.zeros(n+1)
    sims[0] = S0
    for i in range(n):
        sims[i+1] = sims[i] * (1+ np.random.normal(r_dt, s_sqrt_dt))
    return sims    
```
